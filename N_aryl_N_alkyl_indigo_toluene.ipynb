```python
!pip install pandas
!pip install sklearn
!pip install itertools
!pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from itertools import combinations
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("N_aryl_N_alkyl_indigo.csv")
df
```

```python
parm = ['L_C1-C2', 'Occupancy_C1-O', 'Energy_C1-O', 'LUMO', 'HOMO', 'dipole moment', 'polarizability', 'Ster_L', 'Ster_B1', 'Ster_B5', 'Ar_NBO_C1', 'Ar_NBO_C2', 'Ar_NBO_=O', 'Ar_NBO_-O', 'Ar_I_C=O', 'Ar_v_C=O']
df_x = df[parm]
scaler = StandardScaler()
df[parm] = scaler.fit_transform(df_x)
df
```

```python
for i in range (1, 5):
  for c in combinations(parm, i):
    x = df[list(c)]
    y = df['ln(kobs)_toluene']
    regr = linear_model.LinearRegression()
    regr.fit(x,y)
    y_pred = regr.predict(x)
    r2 = regr.score(x,y)
    if r2 > 0.80:
      print(list(c))
      print("rmse =", mean_squared_error(y, y_pred))
      print("r2 =", regr.score(x,y))
      print("")
```

```python
x2 = df[['L_C1-C2', 'LUMO', 'Ar_NBO_C2', 'Ar_NBO_=O']]
y2 = df['ln(kobs)_toluene']
regr = linear_model.LinearRegression()
regr.fit(x2,y2)
y_pred2 = regr.predict(x2)
r2_2 = regr.score(x2,y2)
print("rmse =", mean_squared_error(y2, y_pred2))
print("r2 =", regr.score(x2,y2))
print(regr.coef_)
print(regr.intercept_)
```

```python
import numpy as np
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn import metrics

x_toluene_r = df[['L_C1-C2', 'LUMO', 'Ar_NBO_C2', 'Ar_NBO_=O']]
y_toluene_r = df[['ln(kobs)_toluene']]
x_toluene_array = np.array(x_toluene_r)
y_toluene_array = np.array(y_toluene_r)
loo = LeaveOneOut()
y_toluene_testall = []
y_toluene_predall = []
for train_i, test_i in loo.split(x_toluene_r):
    x_toluene_train, x_toluene_test = x_toluene_array[train_i], x_toluene_array[test_i]
    y_toluene_train, y_toluene_test = y_toluene_array[train_i], y_toluene_array[test_i]
    
    model = LinearRegression()
    model.fit(X = x_toluene_train, y = y_toluene_train)
    y_toluene_pred = model.predict(x_toluene_test)
    
    y_toluene_testall += list(y_toluene_test)
    y_toluene_predall += list(y_toluene_pred)
    
q2 = metrics.r2_score(y_toluene_testall, y_toluene_predall)
ms_error = metrics.mean_squared_error(y_toluene_testall, y_toluene_predall)

print("q2 =", q2, ", rmse =", ms_error)
```

```python
from matplotlib.ticker import FormatStrFormatter

plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 6))

ax.set_facecolor('w')

ax.plot(y_pred2, y_pred2, color='k')
ax.scatter(y2, y_pred2, edgecolor='r', facecolor='r', alpha=0.7)
ax.set_ylabel('Predicted ln(k)', fontsize=18, color='k')
ax.set_xlabel('Experimental ln(k)', fontsize=18, color='k')
ax.spines['bottom'].set_color('k')
ax.spines['left'].set_color('k')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))

fig.text(0.125, 0.96, 'ln(k) = 0.54(L_C1-C2) + 0.13(LUMO) - 0.42(Ar_NBO_C2) - 0.57(Ar_NBO_=O) - 2.34', fontsize = 12)
fig.text(0.60, 0.35, '$R^2= 0.80$', fontsize = 16)
fig.text(0.60, 0.30, 'rmse = 0.048', fontsize = 16)
fig.text(0.60, 0.25, '$Q^2= 0.40$ (LOO)', fontsize = 16)
fig.text(0.60, 0.20, '21 data points', fontsize = 16, style = 'italic')

fig.tight_layout()

plt.savefig('N_aryl_N_alkyl_indigo_toluene.png', bbox_inches='tight')
```
